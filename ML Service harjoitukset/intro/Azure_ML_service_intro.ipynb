Haetaan alkuperäinen, skaalaamaton data verkosta:

```python
Y = df2['Y'].values
gender = df2['SEX']
df_scl = df2.drop(['Y','SEX'],axis=1).apply(lambda x: (x-np.mean(x))/np.std(x))
df_scl.describe().round(2)
```

### *Datan jakaminen*

Ennusmallia kouluttaessa on aina huolehdittava siitä, että käytössä on "riippumatonta" dataa mallin yleistettävyyden testaamiseen. Seuraavassa solussa data jaetaan kahteen osaan:

```python
from sklearn.model_selection import train_test_split

X = df_scl.values
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
data = {
    "train":{"X": X_train, "y": y_train},        
    "test":{"X": X_test, "y": y_test}
}

print('Sample sizes:')
{k:v['X'].shape[0] for k,v in data.items()}
```

## Mallin opetus

Aluksi luodaan eksperimentti-objekti. Jos nimettyä eksperimenttiä ei ole olemassa Azure ML työtilassa, luodaan työtilaan uusi eksperimentti. Muussa tapauksessa objekti linkkautuu olemassa olevaan eksperimenttiin. 

```python
# Get an experiment object from Azure Machine Learning
experiment = Experiment(workspace=ws, name="train-from_local-notebook")
```

```python
experiment
```

Seuraavaksi aloitetaan mallin koulutus ajo:

```python
# Create a run object in the experiment
run = experiment.start_logging()# Log the algorithm parameter alpha to the run

run
```

Tässä ajossa kokeilemme kahta eri mallia, joista kumpaakin voidaan säätää yhdellä hyperparameterillä. Käytetyt parametrit kirjataan käynnissä olevan ajon alle ML eksperimenttiin `run.log()` metodilla:

```python
alpha = 0.05
# Log parameter value:
run.log('alpha', alpha)

n_neighbors = 3
# Log parameter value:
run.log('n_neighbors', n_neighbors)
```

Sitten olemmekin valmiita mallien koulutukseen. Tässä esimerkissä on käytetty regularisoitua lineaariregressiota (`Ridge`) ja kNN-regressiota (`KNeighborsRegressor`), jotka voi halutessaan vaihta toisiin algoritmeihin:

```python
from sklearn.linear_model import Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import r2_score
import joblib

models, preds = {}, {}

# Create sklearn Ridge regression model
models['linear_model'] = Ridge(alpha = alpha)

# Create sklearn KNeighbors regression model
models['kNN_model'] = KNeighborsRegressor(n_neighbors = n_neighbors)

# Fit the models & Calculate predictions:
for m in models.keys():
    models[m].fit(data['train']['X'], data['train']['y'])
    preds[m] = models[m].predict(data['test']['X'])
```

Mielenkiinnosta voidaan visualisoida kuinka hyvin ennustetut ja odotetut arvot vastaavat toisiaan:

```python
plt.scatter(data['test']['y'],preds['kNN_model'],label='kNN')
plt.scatter(data['test']['y'],preds['linear_model'],label='Ridge')
plt.plot([40,300],[40,300],'k-')
plt.legend()
plt.show()
```

Seuraavaksi lasketaan malleille arviointi metriikoita ja kirjataan nekin ajon tietoihin:

```python
# Output the Mean Squared Error & R2-value to the notebook and to the run:
error = {k : mse(data['test']['y'], pred).round(2) for k,pred in preds.items()}
R2 = {k : r2_score(data['test']['y'], pred).round(2) for k,pred in preds.items()}

for m in models.keys():
    print('For model',m)
    print('  Mean Squared Error is', error[m])
    run.log('mse', error[m])
    print('  Coefficient of determination is', R2[m])
    run.log('R2', R2[m])
```

Näyttäisi siltä, että lineaarimalli ennustaa selvästi paremmin vastemuuttujan vaihtelua.

Seuraava steppi on tallentaa koulutetut mallit (tai vain toinen niistä), jotta sitä voidaan tarvittaessa käyttää myöhemmin:

```python
# Save the models to the outputs directory:
for k,m in models.items():
    joblib.dump(value = m, filename='outputs/'+k+'.pkl')
```

Ja lopuksi voimme lopettaa ajon:

```python
# Complete the run
run.complete()
```
